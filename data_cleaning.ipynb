```python
# Count missing values in each column
missing = df.isnull().sum()

# Calculate percentage of missing values
missing_percent = (missing / len(df)) * 100

# Create a table
missing_df = pd.DataFrame({
    "Column": df.columns,
    "Missing Values": missing,
    "Missing %": missing_percent
})

# Sort by highest missing percentage
missing_df = missing_df.sort_values(by="Missing %", ascending=False)

# Display top 20
missing_df.head(20)
```

```python
# List columns you want to remove (only those that actually exist)
columns_to_remove = [
    "Community Board",
    "Census Tract",
    "NTA",
    "Council District",
    "Parent Property Id",
    "Parent Property Name",
    "Property Notes",
    "National Median Reference Property Type",
    "Third Party Certification Date Achieved"
]

# Drop only the columns that are present
df = df.drop(columns=[col for col in columns_to_remove if col in df.columns])

# Check the new shape
print("New dataset shape:", df.shape)
```

```python
print(df.columns.tolist())
```

```python
columns_to_keep = [
    "Property Name",
    "Year Ending",
    "Address 1",
    "City",
    "Postal Code",
    "Borough",
    "Primary Property Type - Self Selected",
    "Largest Property Use Type",
    "Largest Property Use Type - Gross Floor Area (ft²)",
    "Year Built",
    "Number of Buildings",
    "Occupancy",
    "ENERGY STAR Score",
    "Site EUI (kBtu/ft²)",
    "Weather Normalized Site EUI (kBtu/ft²)",
    "Source EUI (kBtu/ft²)",
    "Site Energy Use (kBtu)",
    "Natural Gas Use (kBtu)",
    "Electricity Use - Grid Purchase (kWh)",
    "Total GHG Emissions (Metric Tons CO2e)",
    "Direct GHG Emissions (Metric Tons CO2e)",
    "Indirect GHG Emissions (Metric Tons CO2e)",
    "Property GFA - Self-Reported (ft²)",
    "Latitude",
    "Longitude"
]

# Keep only the selected columns that exist in the dataset
df = df[[col for col in columns_to_keep if col in df.columns]]

# Check the result
print(df.shape)
df.head()
```

```python
# Create a copy so the original data remains unchanged
project_df = df.copy()
```

```python
project_df["Borough"] = project_df["Borough"].fillna("Unknown")
```

```python
project_df = project_df.drop_duplicates()
```

```python
project_df["Year Ending"] = pd.to_datetime(
    project_df["Year Ending"],
    errors="coerce"
)
```

```python
numeric_columns = [
    "Largest Property Use Type - Gross Floor Area (ft²)",
    "ENERGY STAR Score",
    "Site EUI (kBtu/ft²)",
    "Weather Normalized Site EUI (kBtu/ft²)",
    "Source EUI (kBtu/ft²)",
    "Site Energy Use (kBtu)",
    "Natural Gas Use (kBtu)",
    "Electricity Use - Grid Purchase (kWh)",
    "Total GHG Emissions (Metric Tons CO2e)",
    "Direct GHG Emissions (Metric Tons CO2e)",
    "Indirect GHG Emissions (Metric Tons CO2e)",
    "Property GFA - Self-Reported (ft²)"
]

for col in numeric_columns:
    project_df[col] = pd.to_numeric(project_df[col], errors="coerce")
```

```python
project_df.info()
```

```python
project_df[numeric_columns].isnull().sum()
```

```python
project_df = project_df.drop(
    columns=["Property GFA - Self-Reported (ft²)"]
)
```

```python
project_df["Largest Property Use Type - Gross Floor Area (ft²)"] = (
    project_df["Largest Property Use Type - Gross Floor Area (ft²)"]
    .fillna(
        project_df["Largest Property Use Type - Gross Floor Area (ft²)"].median()
    )
)
```

```python
# Display the top 10 most common property types
project_df["Primary Property Type - Self Selected"].value_counts().head(10)
# Plot the top 10 property types
project_df["Primary Property Type - Self Selected"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 Property Types")
plt.xlabel("Property Type")
plt.ylabel("Number of Buildings")
plt.xticks(rotation=90)

plt.show()
```

Insight: The dataset is heavily dominated by Multifamily Housing, with approximately 19,000 records. All other property types have fewer than 2,500 records each, indicating that residential buildings make up the majority of the analyzed properties.

```python
# Count the number of buildings in each borough
project_df["Borough"].value_counts()# Plot the number of buildings in each borough
project_df["Borough"].value_counts().plot(kind="bar")

plt.title("Number of Buildings by Borough")
plt.xlabel("Borough")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.show()
```

Manhattan contains the largest share of benchmarked properties in the dataset, followed by Brooklyn, Bronx, and Queens. Staten Island and records with unknown borough information account for only a small proportion of the data

```python
# Display the 10 buildings with the highest site energy use
project_df.nlargest(10, "Site Energy Use (kBtu)")[
    ["Property Name", "Site Energy Use (kBtu)"]
]
```

A small number of facilities exhibit exceptionally high energy consumption and may represent large institutional campuses or outliers. These values can significantly influence averages and visualizations.

```python
# Displays statistics for ENERGY STAR Score
project_df["ENERGY STAR Score"].describe()

# Plots the distribution of ENERGY STAR Scores
project_df["ENERGY STAR Score"].plot(kind="hist", bins=20)

plt.title("Distribution of ENERGY STAR Scores")
plt.xlabel("ENERGY STAR Score")
plt.ylabel("Number of Buildings")

plt.show()

# Calculates average ENERGY STAR Score for each property type
project_df.groupby(
    "Primary Property Type - Self Selected"
)["ENERGY STAR Score"].mean().sort_values(ascending=False).head(10)
```

The distribution of ENERGY STAR Scores is skewed toward higher values, with many buildings scoring between 70 and 100. This suggests that a significant portion of benchmarked properties demonstrate relatively strong energy performance.
Among the property types with available ratings, Data Centers achieved the highest average ENERGY STAR Score, followed by educational institutions such as Pre-schools and Colleges. Offices and hospitals also performed relatively well, with average scores above 68

```python
# Plot Site EUI after filtering extreme values for better visualization
project_df[project_df["Site EUI (kBtu/ft²)"] < 500]["Site EUI (kBtu/ft²)"].plot(
    kind="hist", bins=30
)

plt.title("Site EUI Distribution (Values < 500)")
plt.xlabel("Site EUI (kBtu/ft²)")
plt.ylabel("Number of Buildings")

plt.show()
```

After excluding extreme outliers for visualization, most buildings exhibit Site EUI values between approximately 40 and 120 kBtu/ft². The distribution is right-skewed, with a relatively small number of buildings consuming substantially more energy per square foot.

```python
# Show the top 10 buildings with the highest GHG emissions
project_df.nlargest(
    10,
    "Total GHG Emissions (Metric Tons CO2e)"
)[["Property Name", "Total GHG Emissions (Metric Tons CO2e)"]]
```

A small number of facilities account for disproportionately high greenhouse gas emissions. These include large campuses, residential complexes, and institutional properties. Such outliers can significantly influence aggregate statistics and should be interpreted separately from typical buildings.

Observation: Two St. Barnabas records have identical exceptionally high values. This may indicate related reporting entities or duplicated aggregate reporting rather than independent facilities.

```python
# Filter extreme values only for better visualization

plt.figure(figsize=(8, 6))

plt.scatter(
    project_df[
        (project_df["Site Energy Use (kBtu)"] < 100000000)
        & (project_df["Total GHG Emissions (Metric Tons CO2e)"] < 10000)
    ]["Site Energy Use (kBtu)"],
    project_df[
        (project_df["Site Energy Use (kBtu)"] < 100000000)
        & (project_df["Total GHG Emissions (Metric Tons CO2e)"] < 10000)
    ]["Total GHG Emissions (Metric Tons CO2e)"],
    alpha=0.5
)

plt.title("Energy Use vs GHG Emissions (Filtered)")
plt.xlabel("Site Energy Use (kBtu)")
plt.ylabel("Total GHG Emissions (Metric Tons CO2e)")

plt.show()
```

Buildings with higher site energy consumption generally produce higher greenhouse gas emissions, indicating a strong positive relationship between energy use and environmental impact.
Although the overall trend is linear, a few outlier buildings exhibit unusual emission patterns and warrant additional investigation.
